# Value Tech - Demonstracao Sintetica

Notebook de exemplo para calcular um score simples de prioridade tecnologica.

```python
import pandas as pd
df = pd.read_csv('../data/tecnologias_exemplo.csv')
df['score_prioridade'] = (df['trl'] * 8 + df['potencial_mercado'] * 0.35 + df['impacto_estimado'] * 0.30 + df['aderencia_estrategica'] * 0.25 - df['risco_tecnico'] * 0.15 - df['risco_regulatorio'] * 0.10)
df.sort_values('score_prioridade', ascending=False)[['tecnologia','setor','score_prioridade']]
```
